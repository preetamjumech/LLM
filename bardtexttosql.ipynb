```python
path='/content/nasdaq_screener_1681283576445.csv'
```

```python
import pandas as pd
company=pd.read_csv(path)
```

```python
def getquery(table,columns,query):
  k=bard.get_answer(f"A table {table} has columns {columns}.Write a Sql query to {query}")
  q=k['content']
  ans=q.split('\n\n')[1][7:-4]
  return ans
```

```python
from sqlalchemy import create_engine, text
def getResult(dataset, query, tableName):
  engine = create_engine('sqlite://',
					echo = False)
  company.to_sql(tableName,
			con = engine)
  with engine.connect() as conn:
    result = conn.execute(text(query))
  return result
```

```python
def getSqlQueryFromQuestion(userQuery, tableName,columns):
  y=getquery(tableName,columns,userQuery)
  #print(y)
  query = y
  print("SQL query returned", query)
  print()
  results = getResult(company, query, tableName)
  for row in results:
    print(row)
```
