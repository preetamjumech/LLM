```python
from litellm import completion

response = completion(
    model="ollama/llama3.2:1b",
    messages=[{ "content": "Write a fastapi code for simple sentiment analysis.","role": "user"}],
    api_base="http://localhost:11434"
)
print(response)
```

```python
formatted_result = response['choices'][0]['message']['content']
formatted_result
```
