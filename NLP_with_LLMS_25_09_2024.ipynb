# Sentence Similarity using Tokenizer and Model

```python
from transformers import AutoTokenizer, TFAutoModel
import numpy as np
import tensorflow as tf
from sklearn.metrics.pairwise import cosine_similarity

MODEL = "sentence-transformers/all-mpnet-base-v2"
MODEL
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL)
tokenizer
```

```python
model = TFAutoModel.from_pretrained(MODEL, from_pt = True)
model
```

```python
sentence1 = "I love cricket"
sentence2 = "Your hobby is travelling"


tokenized_input = tokenizer(
    [sentence1, sentence2],
    padding = True,
    truncation = True,
    return_tensors = "tf"
)

tokenized_input
```

```python
output = model(**tokenized_input)
output
```

```python
last_hidden_state = output.last_hidden_state
last_hidden_state
```

```python
last_hidden_state.dtype
```

```python
mask = tf.cast(tokenized_input["attention_mask"], last_hidden_state.dtype)
mask
```

```python
tokenized_input["attention_mask"].dtype
```

```python
mask.dtype
```

```python
expanded_mask = tf.expand_dims(mask, -1)
expanded_mask
```

```python
masked_embeddings = last_hidden_state * expanded_mask
masked_embeddings
```

```python
summed = tf.reduce_sum(masked_embeddings, axis = 1)
summed
```

```python
summed.shape
```

```python
summed_mask = tf.clip_by_value(tf.reduce_sum(mask, axis = 1),
                               clip_value_min = 1e-9,
                               clip_value_max= tf.reduce_max(mask, axis = 1))

summed_mask
```

```python
tf.reduce_max(mask, axis = 1)
```

```python
tf.expand_dims(summed_mask, axis = -1)
```

```python
mean_pooled = summed/ tf.expand_dims(summed_mask, axis = -1)
mean_pooled
```

```python
mean_pooled = mean_pooled.numpy()
similarity = cosine_similarity([mean_pooled[0], mean_pooled[1]])
similarity
```
