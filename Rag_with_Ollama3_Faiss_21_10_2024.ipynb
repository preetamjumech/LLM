```python
import os
from langchain_community.llms import Ollama
from langchain.document_loaders import UnstructuredFileLoader
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
```

```python
llm = Ollama(model="llama3:instruct",
             temperature = 0)
```

```python
loader = UnstructuredFileLoader("/content/gastroesophageal-surgery-case-scenario-stomach-cancer-mullen.pdf")
documents = loader.load()
```

```python
text_splitter = CharacterTextSplitter(separator = "/n", chunk_size=1000, chunk_overlap=250)
texts = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
knowledge_base = FAISS.from_documents(texts, embeddings)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm,
                                       retriever = knowledge_base.as_retriever())
```

```python
question = "What are the types of diseases she has been facing?"

response = qa_chain.invoke({"query":question})

print(response["result"])
```
